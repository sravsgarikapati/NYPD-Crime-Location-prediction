```python
df.corr(method='pearson')
```

The pandas corr() method does not tell us the p-values. To show the p-value, we can use the scipy.stats.pearsonr() method.

```python
import scipy as sp
```

```python
sp.stats.pearsonr(df['KY_CD'],df['JURISDICTION_CODE'])
```

The correlation between Classification code of the incident and Jurisdiction is 3.44. P-value=0.99 means that the correlation is not statistically significant.

```python
sp.stats.pearsonr(df['ADDR_PCT_CD'],df['KY_CD'])
```

The correlation between Latitude and X-coordinate of Newyork plane coordinate system is statistically significant (p-value < 0.05)

## Correlation Heatmap

We can use seaborn module to create correlation heatmap, which helps us to easily detect high and low pairwise correlations in a dataset.

```python
# Compute and plot the pair-wise correlations
sns.heatmap(df.corr()) 
```

```python
mask = np.zeros_like(df.corr())
mask[np.triu_indices_from(mask)] = True
sns.heatmap(df.corr(), cmap='RdYlGn', vmax=1.0, vmin=-1.0 , mask = mask, linewidths=2)
plt.yticks(rotation=0)
plt.xticks(rotation=45)
plt.savefig('correlation_heatmap.png',dpi=500,bbox_inches='tight')
```
